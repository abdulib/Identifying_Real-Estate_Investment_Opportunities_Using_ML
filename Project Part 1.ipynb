```python
# Looks at type of homes and filter for only SINGLE_FAMILY, CONDO,  and TOWNHOUSE
sold_df['hdpData.homeInfo.homeType'].value_counts()
```

```python
options = ['SINGLE_FAMILY', 'CONDO', 'TOWNHOUSE'] 
    
# selecting rows based on condition 
sold_df = sold_df[sold_df['hdpData.homeInfo.homeType'].isin(options)] 
```

```python
#replace NaN zestimates with price
sold_df['zestimate'] = sold_df['zestimate'].fillna(sold_df['unformattedPrice'])
sold_df['hdpData.homeInfo.taxAssessedValue'] = sold_df['hdpData.homeInfo.taxAssessedValue'].fillna(sold_df['unformattedPrice'])

#rplace nan of rent Zestimate with median rent zestimate
sold_df['hdpData.homeInfo.rentZestimate'] = sold_df['hdpData.homeInfo.rentZestimate'].fillna(sold_df['hdpData.homeInfo.rentZestimate'].median())

# Convert acres to sqft (1 acre * 43560)
sold_df['LotSize (sqft)'] = np.where(sold_df['hdpData.homeInfo.lotAreaUnit'] =='acres',\
                              sold_df['hdpData.homeInfo.lotAreaValue'] * 43560, \
                              sold_df['hdpData.homeInfo.lotAreaValue']
                             )


# sold_df
```

```python
# clean date sold column
sold_df['DateSold'] = sold_df['variableData.text'].str.replace('Sold ', '')

# drop uneeded columns
sold_df.drop(columns=['variableData.text', 'hdpData.homeInfo.lotAreaUnit'], axis=1, inplace=True)
```

```python
sold_df.replace('', np.nan, inplace=True)
```

### Parameter Definitions

- zpid: Unique ID of each house 
- SalePrice: The price the house was sold
- beds: Number of beds in home
- baths: number of baths in home
- area: Size of home in square feet
- latitude: Latitude coordinates of house
- longitude: Longitude coordinates of house
- zestimate - Zillows estimate of house value
- taxAssessedValue: Tax authority valuation of home
- DaysOnZillow - Number of days house was on zillow
- rentZestimate - Zillow estimate of rent 
- lotAreaValue - value of the land house is bulit
- LotSize (sqft) - Size of the land house is bulit in square feet
- DateSold- Date home was sold
- homeType_CONDO: Indicator if home is a condo 
- homeType_SINGLE_FAMILY:  Indicator if home is a Single Family Home 
- homeType_TOWNHOUSE:  Indicator if home is a Townhome
